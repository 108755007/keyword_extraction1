```python
def add_words(self, words=None):
    path = f'{self.ROOT_DIR}/jieba_based/add_words.txt'
    if words == None: ## use words in add_words.txt
        # words = [line.rstrip('\n') for line in open(path, 'r', encoding='utf-8').readlines()]
        words = self.read_file(path)
    elif type(words) == 'str':
            words = [words]
    # if type(words) == 'str':
        #     words = [words]
    for word in words:
        jieba.add_word(word)
        jieba.suggest_freq(word, tune=True)
    self.words_added = words
    return words
```

```python
##########讀取 txt檔####################
from definitions import ROOT_DIR
print(ROOT_DIR)
path = f'{ROOT_DIR}/jieba_based/add_words.txt'
def read_file(path):   
    words = [line.rstrip('\n') for line in open(path, 'r', encoding='utf-8').readlines()]
    return words
```

```python
########讀取關鍵字 
web_id_with_hash_tag = ['ctnews', 'mirrormedia', 'upmedia', 'btnet', 'bnext','dailyview', 'moneyweekly', 'nongnong', 'newscts', 'newtalk','setn', 'nownews']
def fetch_all_hashtags(self):
    query = f"""SELECT keywords FROM article_list WHERE keywords!='_' AND keywords!='' AND web_id in ('{"','".join(self.web_id_with_hash_tag)}')"""
    print(query)
    data = DBhelper('dione').ExecuteSelect(query)
    ##strip(去除頭尾空格)####整理數據乾淨#####
    #先轉換成set 把重複的詞給去掉 再轉換回來list
    hashtag_list = list(set([word.strip() for d in data for word in d[0].split(',')]))
    return hashtag_list
```

```python
def fetch_gtrend_keywords(self):
    ###### 一次取 keyword,relatedQueries 會變成 n*2維數據#####
    query = "SELECT keyword, relatedQueries from google_trend_keyword"
    data = MySqlHelper('dione').ExecuteSelect(query)
    keyword_list = []
    ###將數據整理乾淨 ###
    for d in data:
        if d[1] != '':             
            keyword_list += [d[0]] + d[1].split(',')
    return keyword_list
```
